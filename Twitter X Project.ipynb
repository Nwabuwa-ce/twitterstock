# Basic Analysis of Twitter Stock

The dataset I'm exploring is the dataset of Twitter Stock prices in the period between November 2013 and October 2022.
It is comprised of 2264 entries with 2259 non-null entries recorded between the given period.

```python
# Let's start by importing pandas

import pandas as pd
```

```python
# Then read the csv file into a pandas dataframe.

df_x = pd.read_csv("https://raw.githubusercontent.com/Nwabuwa-ce/twitterstock/main/Twitter-Stock-Market-Dataset.csv")
```

```python
# View the head() of the dataset

df_x.head()
```

```python
df_x.info()
```

```python
# Set the Date column to become the index

df_x.set_index('Date', inplace=True)
df_x.head()
```

I explore the dataset with .info() and .shape

```python
df_x.shape
```

You can see that in the .shape code input the number of rows is 2264 but in the .info() code input the number of non-null rows is 2259 which means there are 5 null values in the dataset. 

```python
print(df_x.isnull().sum())
```

I dropped the null values to get a clean dataset

```python
df_x.dropna()
```

```python
#Let's check the 'count' of rows again, using the describe() method
df_x.describe()
```

Now let's do a little visualization of the dataset

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

# Plot `Twitter Stock` Adj Close  against Date
df_x["Adj Close"].plot(ax=ax, label="Twitter Stock", color="blue")

# Label axes
plt.xlabel("Date")
plt.ylabel("Adj Close")


# Add legend
plt.legend();
```

From the graph plotted for Date and Adj Close, you can see that Twitter (now known as X) stock, fell to under 20 USD between late 2015 and late 2017 and then rose to a high in early 2021.

```python
fig, ax = plt.subplots(figsize=(15, 6))

# Plot `Twitter Stock` Volume against Date
df_x["Volume"].plot(ax=ax, label="Twitter Stock", color="red")

# Label axes
plt.xlabel("Date")
plt.ylabel("Volume")


# Add legend
plt.legend();
```

From the graph plotted for Date and Volume we can see that the Volume spikes at regular intervals but the most prominent spikes are in late 2016 and early 2022.

```python
# Calculate daily returns
df_x['Daily_Return'] = df_x['Adj Close'].pct_change()

# Calculate cumulative returns
df_x['Cumulative_Return'] = (1 + df_x['Daily_Return']).cumprod() - 1

df_x.head()
```

```python
# Plot daily returns
df_x['Daily_Return'].plot(figsize=(15, 6), title=f'Twitter Daily Returns')
plt.show()
```

```python
# 50-day moving average
df_x['MA_50'] = df_x['Adj Close'].rolling(window=50).mean()

# Plot stock prices with 50-day moving average
df_x[['Adj Close', 'MA_50']].plot(figsize=(15, 6), title=f'Twitter Stock with 50-day Moving Average')
plt.show()
```
